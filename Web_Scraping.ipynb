procedemos a limpiar y renombrar columnas del dataframe

```python
def clean_data(df_poblacion):
    # Drop column: 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]​.7'
    df_poblacion = df_poblacion.drop(columns=['Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.7'])
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b' to 'puesto (2022)'
    df_poblacion = df_poblacion.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b': 'puesto (2022)'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.1' to 'Comuna'
    df_poblacion = df_poblacion.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.1': 'Comuna'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.2' to 'CENSO 1991'
    df_poblacion = df_poblacion.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.2': 'CENSO 1991'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.3' to 'CENSO 2001'
    df_poblacion = df_poblacion.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.3': 'CENSO 2001'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.4' to 'CENSO 2010'
    df_poblacion = df_poblacion.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.4': 'CENSO 2010'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.5' to 'CENSO 2022'
    df_poblacion = df_poblacion.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.5': 'CENSO 2022'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.6' to 'CRECIMIENTO 2010-2022'
    df_poblacion = df_poblacion.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.6': 'CRECIMIENTO 2010-2022'})
    # Drop column: 'CENSO 2001'
    df_poblacion = df_poblacion.drop(columns=['CENSO 2001'])
    # Drop column: 'CENSO 1991'
    df_poblacion = df_poblacion.drop(columns=['CENSO 1991'])
    return df_poblacion

df_poblacion = clean_data(df_poblacion.copy())
df_poblacion.head()
```

```python
df_poblacion.info()
```

procedemos a calcular la tasa de crecimiento anual para poder tener los datos de 2021 sobre el indice de crecimiento y poblacion total segun los resultados del censo llevado a cabo en 2022

```python
# eliminamos la primer fila que contiene el titulo de las columnas
df_poblacion = df_poblacion.drop(df_poblacion.index[0])

# convertimos la columna 'CENSO 2022' a float
df_poblacion['CENSO 2022'] = df_poblacion['CENSO 2022'].str.replace('\xa0', '').astype(float)

# poblacion total de 2022
censo_2022 = 3120612

# tasa de crecimiento total entre 2010 y 2022
tasa_crecimiento_total = 7.97 / 100

# calculamos la tasa de crecimiento anual
tasa_crecimiento_anual = (1 + tasa_crecimiento_total) ** (1 / 12) - 1

# creamos una nueva columna con los valores estimados para 2021
df_poblacion['CENSO 2021'] = df_poblacion['CENSO 2022'] / (1 + tasa_crecimiento_anual)

# cambiamos tipo de dato de la columna
df_poblacion['CENSO 2021']= df_poblacion['CENSO 2021'].astype(int)

df_poblacion.head()
```

podemos notar que ya tenemos el censo y la poblacion de las comunas para el año 2021, pero necesitamos de los datos de los demas años. con este proposito, vamos a realizar un calculo basado en la columna 'CRECIMIENTO 2010-2022', que indica el indice de crecimiento porcentaje. este factor sera esencial para el desarrollo de nuestro codigo

```python
# datos de población de 2010 y tasas de crecimiento para nuestro calculo 
data = {
    'Comuna': ['Comuna 13', 'Comuna 14', 'Comuna 12', 'Comuna 4', 'Comuna 1', 'Comuna 7', 'Comuna 8', 'Comuna 6', 'Comuna 11', 'Comuna 15', 'Comuna 3', 'Comuna 5', 'Comuna 10', 'Comuna 9', 'Comuna 2'],
    'Poblacion 2010': [231331, 225970, 200116, 218245, 205886, 220591, 187237, 176076, 189832, 182574, 187537, 179005, 166022, 161797, 157932],
    'Crecimiento 2010-2022': [14.64, 10.19, 18.07, 5.81, 8.44, -1.70, 9.40, 15.73, 7.19, 8.39, 4.22, 8.29, 3.47, 4.78, 0.27]
}

df_poblacion= pd.DataFrame(data)

# convertimos tasas de crecimiento en factores de crecimiento
df_poblacion['Factor de crecimiento']= 1 + df_poblacion['Crecimiento 2010-2022'] / 100

# calculamos la poblacion por cada año
for year in range(2016, 2022):
    n= year - 2010
    df_poblacion[f'Poblacion {year}']= df_poblacion['Poblacion 2010'] * df_poblacion['Factor de crecimiento']**(n/12)

df_poblacion
```

cambiamos tipo de dato de las columnas especificas a int64

```python
# columnas de años a modificar
columnas_a_modificar= ['Poblacion 2016', 'Poblacion 2017', 'Poblacion 2018', 'Poblacion 2019', 'Poblacion 2020', 'Poblacion 2021' ]

# cambiamos tipo de dato a int64
for columna in columnas_a_modificar:
    df_poblacion[columna]= df_poblacion[columna].astype('int64')
```
